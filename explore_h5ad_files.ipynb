```python
import os
import scanpy as sc
import pandas as pd
from scipy.io import mmwrite

# Define output directory
output_dir = '/Users/monicapitu/Desktop/Phenomics/MTX/HUMAN/external_madissoon_natgenet_2023_36543915/FROM_CELLXGENE/ORIGINAL_FILES'

# Load your AnnData object
adata = sc.read_h5ad('lung_5loc_sc_sn_raw_counts_cellxgene.h5ad')

# List of sample names (assuming your sample names are in adata.obs['Sample'])
sample_names = adata.obs['Sample'].unique()

# Function to write MTX files
def write_mtx(adata, sample_name):
    outdir = os.path.join(output_dir, sample_name)
    
    # Create all necessary directories
    os.makedirs(outdir, exist_ok=True)

    # Extract the data for the specific sample
    sample_adata = adata[adata.obs['Sample'] == sample_name]

    # Prepare the matrix and metadata
    mtx = sample_adata.X.T
    print(f'Matrix type: {type(mtx)}')  # Check the type of the matrix

    barcodes = sample_adata.obs.index.to_series()
    features = pd.DataFrame(sample_adata.var.index, columns=['gene'])

    # Convert the sparse matrix to CSR format if necessary
    if not hasattr(mtx, 'tocsc'):  # Check if mtx is not already a sparse matrix
        mtx = mtx.tocsr()

    # File paths
    matrix_fp = os.path.join(outdir, 'matrix.mtx')
    barcodes_fp = os.path.join(outdir, 'barcodes.tsv.gz')
    features_fp = os.path.join(outdir, 'features.tsv.gz')

    # Write MTX file with the appropriate field type
    mmwrite(matrix_fp, mtx, field='integer')  # Use 'integer' or 'real' based on your data

    # Write barcodes and features
    barcodes.to_csv(barcodes_fp, sep="\t", header=False, index=False, compression='gzip')
    features.to_csv(features_fp, sep="\t", header=False, index=False, compression='gzip')

# Iterate through each sample and call the function
for sample_name in sample_names:
    print(f'Processing sample: {sample_name}')
    write_mtx(adata, sample_name)

print('All samples processed and files saved!')
```
